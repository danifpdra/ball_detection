```python
import matplotlib.pyplot as plt
import os
import numpy as np
import statistics 
import math  
```

```python
txt_file = os.path.join("/home/danielarato/ball_detection/Datasets/train/", "IR_camera.txt")
f = open(txt_file, "r")
ir=[]

for line in f: 
    timestamp=line[65:88-4]
    timestamp=float(timestamp)*10**-3
    print(timestamp)
    ir.append(int(timestamp))
print(ir)
```

```python
txt_file_rgb = os.path.join("/home/danielarato/ball_detection/Datasets/train/", "RGB_camera.txt")
f_rgb = open(txt_file_rgb, "r")
rgb=[]


for line_rgb in f_rgb: 
    timestamp_rgb=line_rgb[61:84-4]
    timestamp_rgb=float(timestamp_rgb)*10**-3
    print(timestamp_rgb)
#     print(timestamp_rgb)
    rgb.append(int(timestamp_rgb))
    
    
```

```python
n=min(len(ir),len(rgb))
# print (n_ir,n_rgb)
# print (rgb)
print(timestamp_rgb)
```

```python
plt.figure(figsize=(30,20))
plt.scatter(range(0,n), ir[0:n],color= "red",marker= "*", s=10)
plt.scatter(range(0,n),rgb[0:n],color= "blue", marker= "*", s=10)



# naming the x-axis 
plt.xlabel('Frame number') 
# naming the y-axis 
plt.ylabel('Timestamp (ns)') 
# plot title 
plt.title('Timestamp analysis!')

plt.show()
```

```python
desvio=0

for i in range(0,n):
    desvio+=((ir[i]-rgb[i])*10**-3)**2
    print(ir[i], rgb[i],ir[i]-rgb[i],"\n")

desvio=math.sqrt(abs(desvio)/n)
print(desvio)
```
